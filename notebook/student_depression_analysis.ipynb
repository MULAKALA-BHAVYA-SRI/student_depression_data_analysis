This dataset examines the connection between mental health and various demographic, academic, and lifestyle factors. It includes attributes such as gender, age, academic pressure, study satisfaction, sleep duration, dietary habits, study hours, financial stress, family history of mental illness, depression, and suicidal thoughts. The dataset enables an exploration of how factors like sleep quality, diet, and academic workload impact mental well-being, offering potential applications in identifying patterns of mental health risks and promoting preventive strategies for improved mental health outcomes among different demographic groups.

## Data Visualization

In Data Analysis, visualization is very important. Raw text or numbers can not speak much but figures definitely gives more insights about data.
The visualization can be done by using the python packages like ``numpy``, ``pandas``, ``matplotlib``, ``seaborn``.

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('dark')
#sns.set(style='dark')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (10, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
sd_df
```

#### 1. Gender wise distirbution of students through ``bar`` plot
**Purpose**: To visualize the distribution of gender in the dataset.

```python
print(sd_df["Gender"].value_counts())
sns.countplot(data=sd_df, x='Gender', color = 'skyblue', edgecolor='black')
plt.title('Gender Distribution')
plt.show()
```

**Insight**: Here by using bar plot we can easily identify the number of male vs. female in the dataset. We can see that the number of male are 267 and the number of female are 235.

#### 2. Gender-wise Distribution of Students through ``pie`` chart

```python
print(sd_df["Gender"].value_counts())
sd_df["Gender"].value_counts().plot(kind="pie")
plt.title('Gender-wise Distribution of Students')
plt.show()
```

#### 4. Histogram of Age Distribution
**Purpose**: To understand the distribution of age, helps to identify age groups with more participants.

```python
plt.hist(sd_df['Age'], bins=15, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

**Insights**:The histogram shows the distribution of ages, indicating which age groups are more frequent in the dataset.

#### 7. Clustered Bar Plot of Financial Stress by Gender
Allows you to compare how financial stress varies across gender groups.

```python
sns.countplot(data=sd_df, x='Financial_Stress', hue='Gender', color='r')
plt.title('Financial Stress by Gender')
plt.show()
```

The clustered bar plot shows the distribution of financial stress across different gender categories. Here we can see that the Male are facing Financial Stress over the Female.

#### 8. Heatmap of Correlation Matrix
**Purpose**: To understand the correlation between numerical features.                
**Insight**: Identifies highly correlated variables that may need special attention in modeling.


```python
numeric_df = sd_df.select_dtypes(include=[np.number])

corr = numeric_df.corr()

# Plot the heatmap
sns.heatmap(corr, annot=True, cmap='Purples', linewidths=.5) #, fmt='.2f'
plt.title('Correlation Matrix')
plt.show()
```

The heatmap reveals the strength of correlations between numeric features. Here we can see that there are positive correlations between Academic_Pressure and Age, Academic_Pressure and Study_Hours, Financial_Stress and Study_Hours, and so on.

#### 9. Pairplot of Age, Academic Pressure, and Study Satisfaction
**Purpose**: To explore the relationships between multiple numerical features.            
**Insight**: Helps visualize how features relate to each other.

```python
# plot the pairplot 
sns.pairplot(sd_df[['Age', 'Academic_Pressure', 'Study_Satisfaction']])
plt.show()
```

The pairplot shows scatter plots and correlations between Age, Academic Pressure, and Study Satisfaction.

#### 10. Joint Plot of Study Hours vs. Academic Pressure
**Purpose**: To explore the relationship between study hours and academic pressure.        
**Insight**: Shows if there's a correlation between the amount of time spent studying and the level of academic pressure.

```python
sns.jointplot(data=sd_df, x='Study_Hours', y='Academic_Pressure')
plt.show()
```
